## Gesamtpreis der Mahlzeit

```python
df_fleisch_zutaten["gesamtpreis"] = df_fleisch_zutaten["menge_in_kg"] * df_fleisch_zutaten["preis_pro_kg"]
df_fleisch_zutaten
```

```python
df_vegan_zutaten["gesamtpreis"] = df_vegan_zutaten["menge_in_kg"] * df_vegan_zutaten["preis_pro_kg"]
df_vegan_zutaten
```

```python
gesamtpreis_fleisch = df_fleisch_zutaten.sum(axis = 0)["gesamtpreis"].round(2)
gesamtpreis_vegan = df_vegan_zutaten.sum(axis = 0)["gesamtpreis"].round(2)
print("gesamtpreis_fleisch = " + str(df_fleisch_zutaten.sum(axis = 0)["gesamtpreis"].round(2)) + " €"  )
print("gesamtpreis_vegan = " + str(df_vegan_zutaten.sum(axis = 0)["gesamtpreis"].round(2)) + " €" )
```

## Übersicht für die Teilnehmer

Berechnung des Preises pro Person unter der Annahme, dass das Rezept für je 24 Personen gedacht ist.
Berechnung des Anteils am Lagerpreis unter der Annahme, dass der Lagerbeitrag 25.5 € beträgt.

```python
preis_pro_person_fleisch = (gesamtpreis_fleisch/24).round(2)
preis_pro_person_vegan = (gesamtpreis_vegan/24).round(2)
```

```python
anteil_des_lagerbeitrags_fleisch = ((preis_pro_person_fleisch * 100)/25.5).round(2)
anteil_des_lagerbeitrags_vegan = ((preis_pro_person_vegan * 100)/25.5).round(2)
```

```python
übersicht_für_tn_dict = {
    "mit Fleisch" : {
        "Gesamtpreis in €" : gesamtpreis_fleisch,
        "Preis pro Person in €" : preis_pro_person_fleisch,
        "Anteil des Lagerbeitrags in %" : anteil_des_lagerbeitrags_fleisch
    },
    "vegan" : {
        "Gesamtpreis in €" : gesamtpreis_vegan,
        "Preis pro Person in €" : preis_pro_person_vegan,
        "Anteil des Lagerbeitrags in %" : anteil_des_lagerbeitrags_vegan
    },
}

df_übersicht_für_tn = pd.DataFrame.from_dict(übersicht_für_tn_dict)
df_übersicht_für_tn
```
